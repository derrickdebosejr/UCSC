## Problem 1 

Suppose that 5 cards are selected (without replacement) at random from a regular deck of 52 playing cards:

a. If it is known that at least 2 hearts have been selected, what is the probability that at least 3 hearts have been selected?

b. If it is known that the queen of hearts has been selected, what is the probability that at least 3 hearts have been selected?

### Answer 1

a. If the first 2 cards selected are hearts there are 11 remaining hearts and we want to find the probabilty that we draw 1,2 or, 3 hearts in any order
$$P(hearts \geq 3) = P(hearts = 3) + P(hearts = 4) + P(hearts = 5)\\
P(hearts = 3) = \frac{\binom{11}{1} \binom{39}{2}}{\binom{50}{3}}\\
P(hearts = 4) = \frac{\binom{11}{2} \binom{39}{1}}{\binom{50}{3}}\\
P(hearts = 5) = \frac{\binom{11}{3} \binom{39}{0}}{\binom{50}{3}}\\
P(hearts \geq 3) = \frac{741(11)}{19600} + \frac{55(39)}{19600} + \frac{165}{19600}\\
P(hearts \geq 3) = 0.5337244897$$

b. If the first card selected was a queen of hearts there are 12 remaining hearts and we want to find the probabilty we draw 2, 3 or, 4 hearts in any order.
$$P(hearts \geq 3) = P(hearts = 3) + P(hearts = 4) + P(hearts = 5)\\
P(hearts = 3) = \frac{\binom{12}{2} \binom{38}{2}}{\binom{51}{4}}\\
P(hearts = 4) = \frac{\binom{12}{3} \binom{38}{1}}{\binom{51}{4}}\\
P(hearts = 5) = \frac{\binom{12}{4} \binom{38}{0}}{\binom{51}{4}}\\
P(hearts \geq 3) = \frac{66(703)}{249900} + \frac{220(38)}{249900} + \frac{495}{249900}\\
P(hearts \geq 3) = 0.2211004402$$

````python
class Node:
    """
    This class describes a single node contained within a graph. 
    It has the following instannce level attributes:
    
    ID: An integer id for the node i.e. 1
    heuristic_cost: A float value representing the estimated 
                    cost to the goal node
    """    
    def __init__(self, ID, heuristic_cost):
        self.ID = ID
        self.connected_nodes = []
        self.heuristic_cost = heuristic_cost
        
    def __repr__(self):
        ID = self.ID
        hx = self.heuristic_cost
        if len(self.connected_nodes)==0:
            nodes = 'None'
        else:
            nodes = ','.join(str(cn[1].ID) for cn in self.connected_nodes)
        return 'Node:{}\nh(n):{}\nConnected Nodes:{}'.format(ID, hx, nodes)
        
    def set_connected_nodes(self,connected_nodes):
        """
        Adds edges that lead from this node to other nodes:
        
        Parameters:
        - connected_nodes: A list of tuples consisting of (cost, Node), 
                           where 'cost' is a floating point value 
                           indicating the cost to get from this node 
                           to 'Node' and 'Node' is a Node object
        """
        self.connected_nodes = connected_nodes
    
def build_graph():
    """
    Builds the graph to be parsed by the search algorithms.
    Returns: The starting node, which is the entry point into the graph
    """
    ids = range(13)
    coords = [(0,0), (1,1), (1,0), (1,1), (5,2), (3,1), (3,0), 
              (3,-1), (5,1), (4,1), (4,0), (4,-2), (7,0)]
    
    #https://en.wikipedia.org/wiki/Euclidean_distance
    euclidean_distance = lambda x1y1, x2y2: ((x1y1[0]-x2y2[0])**2 +  (x1y1[1]-x2y2[1])**2)**(0.5)
    
    def build_connected_node_list(from_id, to_ids):
        starting_coords = coords[from_id]
        
        connected_nodes = []
        for to_id in to_ids:
            connected_nodes.append((euclidean_distance(starting_coords, coords[to_id]), all_nodes[to_id]))
            
        return connected_nodes
    
    goal_coords = (7,0)
    all_nodes = [Node(_id, euclidean_distance(coord, goal_coords)) for _id, coord in zip(ids, coords)]
    
    all_nodes[8].set_connected_nodes(build_connected_node_list(8, [12]))
    all_nodes[10].set_connected_nodes(build_connected_node_list(10,[12]))
    all_nodes[5].set_connected_nodes(build_connected_node_list(5, [8]))
    all_nodes[6].set_connected_nodes(build_connected_node_list(6, [9, 10]))
    all_nodes[7].set_connected_nodes(build_connected_node_list(7, [11]))
    all_nodes[1].set_connected_nodes(build_connected_node_list(1, [4,5]))
    all_nodes[2].set_connected_nodes(build_connected_node_list(2, [5,6]))
    all_nodes[3].set_connected_nodes(build_connected_node_list(3, [7]))
    all_nodes[0].set_connected_nodes(build_connected_node_list(0, [1,2,3]))
    
    return all_nodes[0]
````

````python
def depth_first_search(starting_node, goal_node):
    """
    This function implements the depth first search algorithm
    
    Parameters:
    - starting_node: The entry node into the graph
    - goal_node: The integer ID of the goal node.
    
    Returns:
    A list containing the visited nodes in order they were visited with starting node
    always being the first node and the goal node always being the last
    """
    assert type(starting_node) is Node, "starting_node is not an Node!"
    visited_nodes_in_order = [starting_node.ID]
    frontier = []
    frontier.append(starting_node.connected_nodes)
    while frontier:
        #base case to check to see if we explored the whole graph
        if frontier == []:
            return []
        #like a stack top
        top= frontier[0]
        if type(top) is tuple:
            exploring = top[1]
            frontier.pop(0)
            frontier = exploring.connected_nodes+frontier
            visited_nodes_in_order.append(exploring.ID)
            if exploring.ID == goal_node:
                return visited_nodes_in_order
        else:
            #top is a list
            exploring = top.pop(0)
            frontier = exploring[1].connected_nodes+top
            visited_nodes_in_order.append(exploring[1].ID)
            if exploring[1].ID == goal_node:
                return visited_nodes_in_order
    return []




def iterative_deepening_depth_first_search(starting_node, goal_node):
    """
    This function implements the iterative deepening depth first search algorithm
    
    Parameters:
    - starting_node: The entry node into the graph
    - goal_node: The integer ID of the goal node.
    
    Returns:
    A list containing the visited node ids in order they were visited with starting node
    always being the first node and the goal node always being the last
    """
    
    visited_nodes_in_order = []
    depth = 0
    #allowed depth to have maximum rather going to infinity
    while depth < 100:
        #recursive_DLS return either an empty list or solution
        visited_nodes_in_order=recursive_DLS(starting_node, goal_node, depth)
        depth = depth+1
        #determines if we have the correct solution
        if goal_node in visited_nodes_in_order:
            return visited_nodes_in_order
    return []

#global variable
cutoff = []
def recursive_DLS(starting_node, goal_node, depth):
    """
    This function implements the recursive depth limited search algorithm
    
    Parameters:
    - starting_node: The entry node into the graph
    - goal_node: The integer ID of the goal node.
    - depth: The limited depth of DFS
    
    Returns:
    A list containing the visited node ids in order they were visited with starting node
    always being the first node and the goal node always being the last
    """
    #get out of recursive call and increase depth to find the answer
    if depth == 0:
        cutoff_occurred = True
        return cutoff
    #determines if we found our goal node
    if starting_node.ID == goal_node:
        cutoff_occurred = False
        cutoff.append(starting_node.ID)
        return cutoff
    else:
        cutoff_occurred = False
        cutoff.append(starting_node.ID)
        for top in starting_node.connected_nodes:
            if type(top) is tuple:
                child = top[1]
                new_depth = depth-1
                recursive_DLS(child,goal_node, new_depth)
                if goal_node in cutoff:
                    return cutoff
                if new_depth == 0:
                    cutoff_occurred=True
                    return cutoff
        if cutoff_occurred == True:
            return cutoff
        else:
            return []

def dist_plus_heuristic(distance, hueristic):
    return distance + hueristic
    
def a_star_search(starting_node, goal_node):
    """
    This function implements the A* search algorithm
    
    Parameters:
    - starting_node: The entry node into the graph
    - goal_node: The integer ID of the goal node.
    
    Returns:
    A list containing the visited node ids in order they were visited with starting node
    always being the first node and the goal node always being the last
    """
    visited_nodes_in_order = [starting_node.ID]
    frontier = []
    frontier.append(starting_node.connected_nodes)
    explored = []
    while frontier:
        #base case to check to see if we explored the whole graph
        if frontier == []:
            return []
        #use a default high minimum to make sure we find minimum f(n)
        minimum = 100.
        #sets minimum
        for a in frontier[0]:
            if dist_plus_heuristic(a[0],a[1].heuristic_cost) < minimum:
                minimum = dist_plus_heuristic(a[0],a[1].heuristic_cost)
        index = 0
        #searches based on minimum
        for a in frontier[0]:
            #how to determine you made it to the end
            if a[1].heuristic_cost == 0:
                visited_nodes_in_order.append(a[1].ID)
                return visited_nodes_in_order
            #exploring based on the minimum found
            if dist_plus_heuristic(a[0],a[1].heuristic_cost) == minimum:
                exploring = a[index]
                visited_nodes_in_order.append(exploring.ID)
                if exploring.ID == goal_node:
                    return visited_nodes_in_order
                frontier= []
                frontier.append(exploring.connected_nodes)
            index = index + 1
    return []
    
````
