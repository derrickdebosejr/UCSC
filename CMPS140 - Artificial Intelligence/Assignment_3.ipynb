## Problem 1

Consider the following constraint satisfaction problem. A linear graph has nodes of the following colors:

- Red
- Yellow
- Green
- Blue
- Violet

Each node has a domain of {1, 2, ..., 9}.

Each node type as the following constraints on its value:

- Red - No contraints
- Yellow - The rightmost digit of of the product of all its neighbors
- Green - The rightmost digit of the sum of all its neighbors 
- Blue - The leftmost digit of the sum of all its neighbors
- Violet - The leftmost digit of the product of all of its neightbors

As a reminder here is the pseudo code for the Min-Conflicts search algorithm:

Complete the function *solve_csp* defined below. You may find some helper functions useful.

```python
#return true if no constraint conflict otherwise returns false
def no_conflict(nodes,arcs,node_values):
    node_length = len(nodes)
    arc_length = len(arcs)
    #need to make sure each node has no conflicts
    for i in range(0,node_length):
        if nodes[i] == 'R':
            #Red has no constraints that need satisfying
            continue
        if nodes[i] == 'Y':
            #Yellow - The rightmost digit of of the product of all its neighbors
            product = 1
            for j in range(0,arc_length):
                if(arcs[j][0]==i  or arcs[j][1] ==i):
                    #determines the node values of the current nodes neighbors
                    if arcs[j][0]==i:
                        product = product * node_values[arcs[j][1]]
                    if arcs[j][1]==i:
                        product = product * node_values[arcs[j][0]]
            product = product%10
            #if its neighbors' values total to product there is no conflict
            if node_values[i]!=product:
                return False
        if nodes[i] == 'G':
            #Green - The rightmost digit of the sum of all its neighbors
            sum_ = 0
            for j in range(0,arc_length):
                if(arcs[j][0]==i  or arcs[j][1] ==i):
                    #determines the node values of the current nodes neighbors
                    if arcs[j][0]==i:
                        sum_ = sum_ + node_values[arcs[j][1]]
                    if arcs[j][1]==i:
                        sum_ = sum_ + node_values[arcs[j][0]]
            sum_ = sum_%10
            #if its neighbors' values total to sum_ there is no conflict
            if node_values[i]!=sum_:
                return False
        if nodes[i] == 'B':
            #Blue - The leftmost digit of the sum of all its neighbors
            sum_ = 0
            for j in range(0,arc_length):
                if(arcs[j][0]==i  or arcs[j][1] ==i):
                    #determines the node values of the current nodes neighbors
                    if arcs[j][0]==i:
                        sum_ = sum_ + node_values[arcs[j][1]]
                    if arcs[j][1]==i:
                        sum_ = sum_ + node_values[arcs[j][0]]
            while sum_ >= 10:
                sum_ = sum_//10
            #if its neighbors' values total to sum_ there is no conflict
            if node_values[i] != sum_:
                return False
        if nodes[i] == 'V':
            #Violet - The leftmost digit of the product of all of its neightbors
            product = 1
            for j in range(0,arc_length):
                if(arcs[j][0]==i  or arcs[j][1] ==i):
                    #determines the node values of the current nodes neighbors
                    if arcs[j][0]==i:
                        product = product * node_values[arcs[j][1]]
                    if arcs[j][1]==i:
                        product = product * node_values[arcs[j][0]]
            while product >= 10:
                product = product//10
            #if its neighbors' values total to product there is no conflict
            if node_values[i] != product:
                return False
    #after checking all the nodes and none returned False then
    #it is true that there are no conflict
    return True



def solve_csp(nodes, arcs, max_steps):
    """
    This function solves the csp using the MinConflicts Search
    Algorithm.

    INPUTS:
    nodes:      a list of letters that indicates what type of node it is,
                the index of the node in the list indicates its id
                letters = {R, Y, G, B, V}
                
    arcs:       a list of tuples that contains two numbers, indicating the 
                IDS of the nodes the arc connects. 
                
    max_steps:  max number of steps to make before giving up

    RETURNS: a list of values for the solution to the CSP where the 
             index of the value corresponds the value for that
             given node.
    """
    from random import randrange
    
    
    for _ in range(max_steps):
        #we want to check 1000 different random numbers
        node_values = [randrange(1,10), randrange(1,10), randrange(1,10), randrange(1,10), randrange(1,10)]
        for _ in range(max_steps):
            #as we try to fix the conflicts. we want to check if there is no conflicts at any state
            if no_conflict(nodes,arcs,node_values):
                return node_values
            else:
                #there is a conflict
                arc_length = len(arcs)
                for i in range(0,len(nodes)):
                    #as we try to fix the conflicts. we want to check if there is no conflicts at any state
                    if no_conflict(nodes,arcs,node_values):
                        return node_values
                    
                    else:
                        if nodes[i] == 'R':
                            if no_conflict(nodes,arcs,node_values):
                                return node_values
                            #change value of red if theres a conflict
                            if node_values[i] == 9:
                                node_values[i] = 1
                            else:
                                node_values[i] = node_values[i] + 1
                        if nodes[i] == 'Y':
                            #Yellow - The rightmost digit of of the product of all its neighbors
                            product = 1
                            for j in range(0,arc_length):
                                if(arcs[j][0]==i  or arcs[j][1] ==i):
                                    #determines the node values of the current nodes neighbors
                                    if arcs[j][0]==i:
                                        product = product * node_values[arcs[j][1]]
                                    if arcs[j][1]==i:
                                        product = product * node_values[arcs[j][0]]

                            product = product%10
                            #checker to see if current color has no conflicts
                            #change value if theres a conflict
                            if node_values[i] != product:
                                if node_values[i] == 9:
                                    node_values[i] = 1
                                else:
                                    node_values[i] = node_values[i] +1
                        if nodes[i] == 'G':
                            #Green - The rightmost digit of the sum of all its neighbors
                            sum_ = 0
                            for j in range(0,arc_length):
                                if(arcs[j][0]==i  or arcs[j][1] ==i):
                                    #determines the node values of the current nodes neighbors
                                    if arcs[j][0]==i:
                                        sum_ = sum_ + node_values[arcs[j][1]]
                                    if arcs[j][1]==i:
                                        sum_ = sum_ + node_values[arcs[j][0]]
                            sum_ = sum_%10
                            #checker to see if current color has no conflicts
                            #change value if theres a conflict
                            if node_values[i] != sum_:
                                if node_values[i] == 9:
                                    node_values[i] = 1
                                else:
                                    node_values[i] = node_values[i] +1
                        if nodes[i] == 'B':
                            #Blue - The leftmost digit of the sum of all its neighbors
                            sum_ = 0
                            for j in range(0,arc_length):
                                if(arcs[j][0]==i  or arcs[j][1] ==i):
                                    #determines the node values of the current nodes neighbors
                                    if arcs[j][0]==i:
                                        sum_ = sum_ + node_values[arcs[j][1]]
                                    if arcs[j][1]==i:
                                        sum_ = sum_ + node_values[arcs[j][0]]
                            while sum_ >= 10:
                                sum_ = sum_//10
                            #checker to see if current color has no conflicts
                            #change value if theres a conflict
                            if node_values[i] != sum_:
                                if node_values[i] == 9:
                                    node_values[i] = 1
                                else:
                                    node_values[i] = node_values[i] +1
                        if nodes[i] == 'V':
                            #Violet - The leftmost digit of the product of all of its neightbors
                            product = 1
                            for j in range(0,arc_length):
                                if(arcs[j][0]==i  or arcs[j][1] ==i):
                                    #determines the node values of the current nodes neighbors
                                    if arcs[j][0]==i:
                                        product = product * node_values[arcs[j][1]]
                                    if arcs[j][1]==i:
                                        product = product * node_values[arcs[j][0]]
                            while product >= 10:
                                product = product//10
                            #checker to see if current color has no conflicts
                            #change value if theres a conflict
                            if node_values[i] != product:
                                if node_values[i] == 9:
                                    node_values[i] = 1
                                else:
                                    node_values[i] = node_values[i] +1
    #if we never found a solution just return the current node values
    return node_values
```

## Problem 2   -  the n rooks problem

Rooks can move any number of squares horizontally or vertically on a chess board. The n rooks problem is to arrange rooks on an $n\times n$ board in such a way that none of the rooks could bump into another by making any of its possible horizontal or vertical moves. 

For this problem, the variables are each column (labeled 0, 1, ... , $n-1$), the the domain consists of each possible row (also labeled 0, 1, ... ,$ n-1$). In each column we place a rook on row 0, row 1, ... , row $n-1$.

For example, if $n=2$ we have only two solutions to this problem:

    R @
    @ R 
    
or

    @ R
    R @ 
   
How many possible solutions are there to this CSP, in terms of n?  
Also, give a simple proof that your answer is correct.

### Answer: There are n! possible solutions to this CSP 
We can place down the first rook in any n number of columns and any n number of rows. Then when we place down the next rook there will now be n-1 rows and columns that rook can be placed in because 1 column and 1 row can no longer be selected because of the first placed down rook constraining the next rooks values. Until we get to the last rook that needs to be placed in and there is only one place that rook can be placed without it conflicting with the already placed down rooks.

```python
# Add the given constraints:
problem.addConstraint(lambda a, b: a**2 == b and (a % 2) == 1, ("x", "y"))

# There are only 5 solutions that satisfy those constraints:
solutions = problem.getSolutions()
print( len(solutions) )
```

### The n-rooks problem revisited

The folling code block shows a solution to this problem using the *constraint* module, with $n=8$.

```python
# The n rooks problem

problem = Problem()
n = 8
cols = range(n)
rows = range(n)
problem.addVariables(cols, rows)
for c1 in cols:
    for c2 in cols:
        if c1 < c2:
            problem.addConstraint(lambda r1, r2: r1 != r2, (c1, c2))

solutions = problem.getSolutions()
len(solutions)
```

## Problem 3  -  The n-queens problem

In chess, a queen can make all the moves of a rook, but also can move along diagonals. As before, our goal is to arrange $n$ queens on an $n\times n$ board so that none of their moves are impeded by the presense of another queen on any row, column, or diagonal.

Modify the $n$ rooks problem to solve the n queens problem. You only need one call to *problem.addConstraint*, as before, since you can use logical 'and's to include as many other constraints in the lambda function as you need.

```python
# The n queens problem

problem = Problem()
n = 8
cols = range(n)
rows = range(n)
problem.addVariables(cols, rows)
for c1 in cols:
    for c2 in cols:
        if c1 < c2:
            # if you subtract the rows and columns of the 2 queens 
            # and they are equal to each other then the queens conflict on the diagonals
            problem.addConstraint(lambda r1, r2, a=c1, b=c2: r1 != r2 and abs(r1 - r2) !=abs(a - b), (c1, c2))
            
solutions = problem.getSolutions()
print(len(solutions))

# The n queens problem

problem = Problem()
n = 11
cols = range(n)
rows = range(n)
problem.addVariables(cols, rows)
for c1 in cols:
    for c2 in cols:
        if c1 < c2:
            # if you subtract the rows and columns of the 2 queens 
            # and they are equal to each other then the queens conflict on the diagonals
            problem.addConstraint(lambda r1, r2, a=c1, b=c2: r1 != r2 and abs(r1 - r2) != abs(a - b), (c1, c2))
            
solutions = problem.getSolutions()
print(len(solutions))
```
